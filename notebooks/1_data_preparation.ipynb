```python
# Import libraries
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

print("✅ Libraries imported!")
```

```python
# Load the dataset
# Make sure the CSV file is in your data folder

df = pd.read_csv('../data/WA_Fn-UseC_-Telco-Customer-Churn.csv')

print("✅ Data loaded successfully!")
print(f"Shape: {df.shape[0]} rows × {df.shape[1]} columns")
```

```python
# Check missing values
print("Missing Values:")
print("=" * 40)
missing = df.isnull().sum()
missing_pct = (df.isnull().sum() / len(df)) * 100

missing_df = pd.DataFrame({
    'Missing Count': missing,
    'Missing %': missing_pct
})

print(missing_df[missing_df['Missing Count'] > 0])
print("\n✅ Missing value check complete!")
```

```python
# Check unique values for each column
print("Unique Values Per Column:")
print("=" * 40)
for col in df.columns:
    print(f"{col:<25} : {df[col].nunique()} unique values")
```

```python
# Check target variable distribution
print("Target Variable: Churn")
print("=" * 40)
churn_counts = df['Churn'].value_counts()
churn_pct = df['Churn'].value_counts(normalize=True) * 100

for val in churn_counts.index:
    print(f"  {val}: {churn_counts[val]} customers ({churn_pct[val]:.1f}%)")

print(f"\n⚠️  Class Imbalance: {churn_pct['No']:.1f}% stayed vs {churn_pct['Yes']:.1f}% churned")
```

```python
# DATA CLEANING

# Step 1: Drop customerID (not useful)
df = df.drop('customerID', axis=1)
print("✅ Dropped customerID column")

# Step 2: Fix TotalCharges (it has spaces, should be numeric)
print(f"\nTotalCharges dtype before: {df['TotalCharges'].dtype}")
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
print(f"TotalCharges dtype after : {df['TotalCharges'].dtype}")

# Step 3: Check how many NaN created
nan_count = df['TotalCharges'].isnull().sum()
print(f"NaN values created: {nan_count}")

# Step 4: Fill NaN with median
df['TotalCharges'].fillna(df['TotalCharges'].median(), inplace=True)
print(f"✅ Filled {nan_count} NaN values with median")

print("\n✅ Data cleaning complete!")
```

```python
# ENCODE CATEGORICAL VARIABLES
from sklearn.preprocessing import LabelEncoder

# Make a copy for encoding
df_encoded = df.copy()

# Get all categorical columns
categorical_cols = df_encoded.select_dtypes(include=['object']).columns.tolist()
print("Categorical columns to encode:")
print(categorical_cols)

# Remove target from list
if 'Churn' in categorical_cols:
    categorical_cols.remove('Churn')

# Encode each column
label_encoders = {}
for col in categorical_cols:
    le = LabelEncoder()
    df_encoded[col] = le.fit_transform(df_encoded[col])
    label_encoders[col] = le
    print(f"  ✅ Encoded: {col}")

# Encode target variable
df_encoded['Churn'] = df_encoded['Churn'].map({'Yes': 1, 'No': 0})
print("  ✅ Encoded: Churn (Yes=1, No=0)")

print("\n✅ Encoding complete!")
```

```python
# View encoded dataframe
print("Encoded DataFrame:")
df_encoded.head()
```

```python
# Save cleaned data
df_encoded.to_csv('../data/cleaned_data.csv', index=False)
print("✅ Cleaned data saved to ../data/cleaned_data.csv")
print(f"Shape: {df_encoded.shape}")
```
